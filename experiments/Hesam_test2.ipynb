```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
import matplotlib as mpl

N = 10000
n = 200
ran_numbs = np.random.normal(0,1,(N,n))



num_array_norm = la.norm(ran_numbs,axis=0)
num_array_ave = la.norm(ran_numbs)/np.sqrt(n)

x = np.linspace(0, 200, num=200)

plt.plot(num_array_norm, label='norms')
plt.ylabel('sum numbers')
plt.plot(x, num_array_ave * np.ones((200)), label='avg')
plt.legend()
```

$$
n = 1000,\quad
x = [x_1,\cdots,x_n]
$$

The Euclidean distance to the origin of $x$ is given by
$$
||x||
\;=\;
\sqrt{\sum_1^n x_i^2}
$$

Now, assume $x_i\sim \mathcal{N}(0,1)$ are i.i.d. standard nomral random variables. Thus, $x$ is a Gaussian random vector and $||x||$ is a well-defined random variable. We have
$$
\mathbb{E}[x] = 0, \quad \mathbb{E}[||x||] = \sqrt{n}
$$
